```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.read_csv('Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv', engine='python')
df2 = pd.read_csv('Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv', engine='python')
df3 = pd.read_csv('Friday-WorkingHours-Morning.pcap_ISCX.csv', engine='python')
df4 = pd.read_csv('Monday-WorkingHours.pcap_ISCX.csv', engine='python')
df5 = pd.read_csv('Thursday-WorkingHours-Afternoon-Infilteration.pcap_ISCX.csv', engine='python')
df6 = pd.read_csv('Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX.csv', engine='python')
df7 = pd.read_csv('Tuesday-WorkingHours.pcap_ISCX.csv', engine='python')
df8 = pd.read_csv('Wednesday-workingHours.pcap_ISCX.csv', engine='python')

frames = [df1, df2, df3, df4, df5, df6, df7, df8]

result = pd.concat(frames)
result.to_csv("cicids2017_COMBINED.csv", sep='\t', encoding='utf-8')
```

```python
df = result
df[' Label'].value_counts()
```

```python
df.columns = df.columns.str.strip()
df = df.drop(columns=['Fwd Header Length.1'])
df.shape
```

```python
df['Label'].unique()
```

```python
nan_counts = df.isna().sum()
def count_infs(col):
    if pd.api.types.is_numeric_dtype(col):
        return np.isinf(col).sum()
    else:
        return 0
inf_counts = df.apply(count_infs)
not_helpful_counts = nan_counts + inf_counts
features_with_issue = not_helpful_counts[not_helpful_counts > 0]
print("Features with at least one NaN or Infinity value:")
print(features_with_issue)
plt.figure(figsize=(15, 7))
not_helpful_counts.plot(kind='bar', color='orange')
plt.title("Not Helpful Values (NaN + Infinity) per Feature")
plt.xlabel("Features")
plt.ylabel("Count of Not Helpful Values")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
df.fillna(0, inplace=True)
df.replace('Infinity', -1, inplace=True)
df[["Flow Bytes/s", "Flow Packets/s"]] = df[["Flow Bytes/s", "Flow Packets/s"]].apply(pd.to_numeric)
df.replace([np.inf, -np.inf, np.nan], -1, inplace=True)
```

```python
nan_counts = df.isna().sum()
inf_counts = df.apply(count_infs)
print("Features with NaNs")
print(nan_counts[nan_counts>0])
print("Features with Infs")
print(inf_counts[inf_counts>0])
```

```python
string_features = list(df.select_dtypes(include=['object']).columns)
string_features.remove('Label')
string_features
```

```python
le = preprocessing.LabelEncoder()
df[string_features] = df[string_features].apply(lambda col: le.fit_transform(col))
df['Label'].value_counts()
```

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE

df_binary = df.copy()

attack_classes = df_binary['Label'].unique().tolist()
attack_classes.remove('BENIGN')
df_binary['Label'] = df_binary['Label'].apply(lambda x: 'Attack' if x != 'BENIGN' else 'BENIGN')

X = df_binary.drop('Label', axis=1)
y = df_binary['Label']

# Undersampling BENIGN
undersampler = RandomUnderSampler(random_state=42)
X_under, y_under = undersampler.fit_resample(X, y)

# Oversampling ATTACK
oversampler = SMOTE(random_state=42)
X_over, y_over = oversampler.fit_resample(X, y)
```
